## Creating a datafarame from stack overflow dataset and analyzing percentage of people knowing Python from different countries

```python
total_resp=df['Country'].value_counts()
total_resp
```

```python
grp=df.groupby(['Country'])
knows_python=grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
knows_python
```

```python
python_df=pd.concat([total_resp,knows_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'total_resp','LanguageWorkedWith':'knows_python'},inplace=True)
```

```python
python_df
```

```python
python_df['%_knows_python']= (python_df['knows_python']/python_df['total_resp'])*100
```

```python
python_df.sort_values(by='%_knows_python',ascending=False).head(50)
```

```python
python_df.head(15)
```

## Percentage of people using Windows Operating System in different countries from our survey data

```python
total_resp=df['Country'].value_counts()
```

```python
grp=df.groupby(['Country'])
windows=grp['OpSys'].apply(lambda x:x.str.contains('Windows').sum())
```

```python
win_df=pd.concat([total_resp,windows],axis='columns',sort=False)
```

```python
win_df.rename(columns={'Country':'total_resp','OpSys':'use_win'},inplace=True)
```

```python
win_df['%_uses_windows']= (win_df['use_win']/win_df['total_resp'])*100
```
